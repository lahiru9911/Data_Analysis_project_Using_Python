```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Load The DataSet

```python
ipl = pd.read_csv('matches.csv')
```

```python
ipl
```

# Analysis

```python
ipl.shape
```

```python
pomc = ipl['player_of_match'].value_counts()[0:10]
```

```python
pom = list(ipl['player_of_match'].value_counts()[0:10].keys())
```

```python
plt.bar(pom,pomc,color='r')
plt.xticks(rotation=90)
plt.show()
```

```python
ipl['result'].value_counts()
```

```python
bat_f_w = ipl[ipl['win_by_runs']!=0]
```

```python
bat_f_w
```

```python
plt.hist(bat_f_w['win_by_runs'])
plt.xlabel('No of Runs that WinBy')
plt.ylabel('No Of Matches')
```

```python
win = bat_f_w['winner'].value_counts()
```

```python
plt.bar(list(win.keys()),win)
plt.xticks(rotation=90)
plt.show()
```

```python
bat_s_w = ipl[ipl['win_by_wickets']!=0]
```

```python
bat_s_w
```

```python
plt.hist(bat_s_w['win_by_wickets'],bins=30)
plt.xlabel('Wickets')
plt.ylabel('No of Match')
plt.show()
```

```python
win = bat_s_w['winner'].value_counts()
```

```python
plt.bar(list(win.keys()),win)
plt.xticks(rotation=90)
plt.show()
```
